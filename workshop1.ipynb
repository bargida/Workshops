Q6.Write a summary of the outcome of data.describe().

Here is a summary of data.describe:

count: The number of non-missing values in each column.
mean: The average value of each column.
std: The standard deviation, which measures the spread or dispersion of the data around the mean.
min: The minimum value in each column.
25%: The 25th percentile (first quartile). 25% of the values in the column are below this value.
50%: The 50th percentile (median). Half of the values in the column are below this value.
75%: The 75th percentile (third quartile). 75% of the values in the column are below this value.
max: The maximum value in each column.
