Histograms are graphical techniques used to represent frequency distribution for a range of value of a single variable by aggregating data into categories called bins or intervals. They are useful to illustrate the shape, the center, as well as the spread of the data.

Boxplots (or box-and-whisker plots) helps in depicting the visually important features of a distribution like the median, the quartiles and some extreme values, which may be outliers. It gives a quick graphic overview of the data.
They also can have other benefits, such as being more robust to outliers than histograms.



4a. What are the extreme values? Are there any outliers(s) exist in this dataset? Explain with example based on the 'ApplicantIncome'?

```python
quartile1 = data['ApplicantIncome'].quantile(0.25)
quartile3 = data['ApplicantIncome'].quantile(0.75)
IQR = quartile3 - quartile1

upper_bound = quartile3 + 1.5 * IQR
lower_bound = quartile1 - 1.5 * IQR


outliers = data[(data['ApplicantIncome'] > upper_bound) | (data['ApplicantIncome'] < lower_bound)]

print(outliers)
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, classes, test_size= .33,
random_state = 8267)
print(x_train.shape, x_test.shape)
```

```python
decisionTree = DecisionTreeClassifier(criterion='entropy')
print(decisionTree)
```

```python
dtc_model = decisionTree.fit(x_train, y_train)
```

```python
importance = dtc_model.feature_importances_
for i,v in enumerate(importance):
  print('Feature: %0d, Score: %.5f' % (i,v))

pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```

```python
prediction = dtc_model.predict(x_test) #prediction stores the predicted targets/classes
```

```python
y_true = le.inverse_transform(y_test["LoanStatus"])
y_pred = le.inverse_transform(prediction)
```

```python
cm = confusion_matrix(y_true, y_pred)
labels = ['Y', 'N']
ConfusionMatrixDisplay(cm, display_labels=labels).plot()
```

```python
print(classification_report(y_true, y_pred))
```

The baseline model had all the features from the normalized dataset. On the other hand, the second model is limited to the top-selected features from the feature importance analysis: 'Credit_History', 'TotalIncome', 'LoanAmount_log', and 'Married'. The assumption here would be that these are the most predictive variables in this model for loan approval.

Comparative Analysis:

The baseline model, with all features, provided a decent accuracy and F1-score. However, the second model, using selected features for training, demonstrated comparable or even better performance. This suggests that a judiciously chosen subset of features can be as effective, if not better than, using all the available features.

The selected features include, but are not limited to, 'Credit_History', which reflects the importance of financial history in loan approval decisions. The inclusion of 'TotalIncome' and 'LoanAmount_log' is in line with the logical consideration of income and loan amount in credit assessment. While 'Married' may have less direct influence, its inclusion may suggest a potential impact on financial stability and loan repayment capability.

Overall, this experiment also demonstrates the potential of feature selection in optimizing model performance. Selecting relevant features judiciously could give comparable or superior results with a less complicated model, thus enhancing generalization.
